## **Basics of Linear Regression**

### **Libraries and Data**
In this example we're going to use a DF with car's info, also we import the libraries needed for having a nice analysis and visualization.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder #One Hot Encoding

dataset = pd.read_csv("./data/Automobile.csv")
```

### **Understanding the data**
After having the data downloaded, now we have to understand the data. Is important to know that the Linear Regression's Algorithm is based on correlations, that's why is highly recommended to start with a chart of the data, using **pairplot** of Seaborn.

```python
import seaborn as sns
sns.set(style='whitegrid', context='notebook')
columns = ['year', 'selling_price', 'km_driven']
sns.pairplot(dataset[columns], height=2.5)
```

After the correlation of the numerical data, we should visualize the patterns and then use a numerical method to understand the current and possible correlation between numerical data.

```python
import numpy as np
cm = np.corrcoef(dataset[columns].values.T)
sns.set(font_scale=1.5)
sns.heatmap(cm, cbar=True, annot=True, yticklabels=columns, xticklabels=columns)
```

### **Model's creation**

```python
from sklearn.preprocessing import StandardScaler #Scaling of Data
from sklearn.linear_model import LinearRegression #Creation of Linear Regression

x = dataset['selling_price'].values.reshape(-1, 1)
y = dataset['km_driven'].values.reshape(-1, 1)

sc_x = StandardScaler()
sc_y = StandardScaler()

x_std = sc_x.fit_transform(x)
y_std = sc_y.fit_transform(y)

slr = LinearRegression()
slr.fit(x_std, y_std)
```

### **Data Analysis**

```python
import matplotlib.pyplot as plt
plt.scatter(x_std, y_std)
plt.plot(x_std, slr.predict(x_std), color='r')
plt.title("Car Prices")
plt.xlabel("Selling Price")
plt.ylabel("Km Driven")
```

### **Generating Prediction Model**

```python
test_price = 80000
price_std = sc_x.transform(np.array((test_price)).reshape(-1, 1))

print(f"Price of a car with 80.000KM is: {sc_y.inverse_transform(slr.predict(price_std))}")
```
