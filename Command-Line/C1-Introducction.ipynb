## 💻**Introducction to the Command Line**🧮

The **Terminal** is the graphic interface that shows us the *Prompt* and simulates the Command Line, this is called **Shell**, this program interprets the commands of the *User* and it executes them on the **Operative System**.

![EmptyShell](./images/C1-EmptyShell.PNG)


Some tipes of **shells** are:
- Bourne Shell
- Bash Shell **(Most used)**
- Z Shell **(Most used)**
- C Shell
- Korn Shell
- Fish Shell
- PowerShell **(Most used)**

### **What is a command?**

A **command** is a program that can be executed from the terminal. This can take parameters and options.
